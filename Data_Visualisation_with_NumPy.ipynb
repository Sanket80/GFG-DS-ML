```python
import numpy as np
import matplotlib.pyplot as plt
```

### 1. Line Plot

##### 1.1) Basic Line Plot

```python
x = np.linspace(0, 10, 100)
y = np.sin(x)
plt.plot(x, y)
plt.xlabel('x')
plt.ylabel('sin(x)')
plt.title('Basic Line Plot')
plt.show()
```

##### 1.2) Custom Line Graph

```python
x = np.linspace(0, 10, 100)
y = np.sin(x)

plt.plot(x, y, color='blue', linewidth=2, linestyle='--', label='sin(x)', alpha = .5)
plt.xlabel('x')
plt.ylabel('sin(x)')
plt.title('Custom Line Graph')
plt.legend()
plt.show()
```

##### 1.3) Line Plot with Multiple Lines

```python
x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.cos(x)

plt.plot(x, y1, color='blue', linewidth=2, linestyle='--', label='sin(x)')
plt.plot(x, y2, color='red', linewidth=2, linestyle='-', label='cos(x)', alpha = .6)
plt.xlabel('x')
plt.ylabel('sin(x), cos(x)')
plt.title('Line Plot with Multiple Lines')
plt.legend()    # to display labels
plt.show()
```

### 2. Scatter Plot

##### 2.1) Basic Scatter Plot

```python
x = np.random.rand(400)
y = np.random.rand(400)

plt.scatter(x, y)
plt.xlabel('x')
plt.ylabel('y')
plt.title('Scatter Plot')
plt.show()
```

##### 2.2) Customized Scatter Plot

```python
x = np.random.rand(400)
y = np.random.rand(400)
z = np.random.rand(400)

plt.scatter(x, y, c=z, cmap='viridis', alpha=0.8)
plt.xlabel('x')
plt.ylabel('y')
plt.title('Scatter Plot')
plt.colorbar()
plt.show()
```

##### 2.3) Scatter Plot with Multiple Classes

```python
x1 = np.random.rand(200) - .5
y1 = np.random.rand(200) - .5
x2 = np.random.rand(200) + .5
y2 = np.random.rand(200) + .5

plt.scatter(x1, y1, c='blue', label='Class 1', alpha=.7)
plt.scatter(x2, y2, c='red', label='Class 2', alpha = .7)
plt.xlabel('x')
plt.ylabel('y')
plt.title('Scatter Plot with Multiple Classes')
plt.legend()
plt.show()
```

##### 2.4) Hexbining Scatter Plot

```python
x = np.random.rand(1000)
y = np.random.rand(1000)

plt.hexbin(x, y, gridsize=20, cmap='viridis')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Hexbin Scatter Plot')
plt.colorbar()
plt.show()
```

### 3.Bar Graph

##### 3.1) Basic Bar Graph

```python
x = np.array(['A', 'B', 'C', 'D'])
y = np.array([3, 1, 4, 2])

plt.bar(x, y)
plt.xlabel('Category')
plt.ylabel('Value')
plt.title('Basic Bar Graph')
plt.show()
```

##### 3.2) Stacked Bar Graph

```python
data = np.random.rand(5, 3)    # 5 random values of 3 categories

plt.bar(np.arange(5), data[:,0], label = 'D-1')
plt.bar(np.arange(5), data[:,1], bottom = data[:,0], label = 'D-2')
plt.bar(np.arange(5), data[:,2], bottom = data[:,0] + data[:,1], label = 'D-3')
plt.xlabel('Category')
plt.ylabel('Value')
plt.title('Stacked Bar Graph')
plt.legend()
plt.show()
```

##### 3.3) Grouped Bar Group

```python
data = np.array([[55,20,30],
                 [40,50,60],
                 [70,80,90]])

width = 0.25
x = np.arange(3)

plt.bar(x, data[0], width, label = 'Group 1')
plt.bar(x + width, data[1], width, label = 'Group 2')
plt.bar(x + 2*width, data[2], width, label = 'Group 3')
plt.xlabel('Category')
plt.ylabel('Value')
plt.title('Grouped Bar Graph')
plt.legend()
plt.show()
```

### 4. Histogram

###### a. Basic Histogram

```python
data = np.random.randn(1000)

plt.hist(data, bins=30, color='blue', alpha=0.7)    # bins = number of bars
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.title('Histogram')
plt.show()
```

###### b. Cummulative Histogram

```python
data = np.random.randn(1000)

# Cumulative adds up the values of the previous bars
plt.hist(data, bins=30, cumulative=True, color='blue', alpha=0.7)
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.title('Cummulative Histogram')
plt.show()
```

###### c.Histogram with Multiple Datasets

```python
data1 = np.random.randn(1000)
data2 = np.random.randn(1000)

plt.hist(data1, bins=30, color='blue', alpha=0.5, label='Dataset 1')
plt.hist(data2, bins=30, color='red', alpha=0.5, label='Dataset 2')
plt.legend()
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.title('Histogram with Multiple Datasets')
plt.show()
```

### 5.Box Plot

```python
data = np.random.randn(100,3)

plt.boxplot(data, labels = ['D-1', 'D-2', 'D-3'], vert = False)
plt.xlabel('Data')
plt.ylabel('Value')
plt.title('Box Plot')
plt.show()
```

### 6. HeatMap

```python
data = np.random.rand(10, 10)

plt.imshow(data, cmap='viridis')
plt.colorbar()
plt.xlabel('x')
plt.ylabel('y')
plt.title('Heatmap')    # HeatMap shows the relationship between two variables
plt.show()
```

### 7. Pie Chart

```python
labels = ['A', 'B', 'C', 'D']
sizes = [15, 30, 45, 10]
explode = (0, 0.1, 0, 0)
colors = ['gold', 'lightskyblue', 'lightpink', 'yellowgreen']

plt.title('Pie Chart')
plt.pie(sizes, explode=explode, labels=labels, colors=colors, autopct='%1.1f%%', shadow=True, startangle=90)
plt.axis('equal')
plt.title('Pie Chart')
plt.show()
```
